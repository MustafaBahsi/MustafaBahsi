İkinci dereceden $ax^2+bx+c=0$ denkleminin terimlerinin $a$, $b$ ve $c$ katsayıları girildiğinde çözümü hakkında bilgi veren kodu yazalım. <br>
* $b^2-4ac$ diskıriminant değerini hesapla
* Diskıriminanat değeri sıfırdan büyükse iki reel kök vardır yazdır.
* Diskıriminanat değeri sıfıra eşitse tek reel kök vardır yazdır.
* Diskıriminanat değeri sıfırdan küçükse reel kök yoktur yazdır.

```python
a = 1
b = 2
c = 1
```

```python
delta = b ** 2 - 4 * a * c
```

```python
if delta > 0:
  print('İki farklı gerçel kök vardır.')
elif delta == 0:
  print('Tek gerçel kök vardır.')
else:
  print('Gerçel kök yoktur.')
```

```python
if delta < 0:
  print('Gerçel kök yoktur.')
else:
  if delta == 0:
    print('Tek gerçel kök vardır.')
  else:
    print('İki farklı gerçel kök vardır.')
```

```python
# katsayıları girilerek 2. dereceden denklemin çözümünü bulan algoritma
a = 1
b = 5
c = 6
delta = b ** 2 - 4 * a * c

if delta > 0:
  print('İki farklı gerçel kök vardır:')
  x1 = (- b + delta ** (1 / 2)) / 2 * a
  x2 = (- b - delta ** (1 / 2)) / 2 * a
  print(x1 , x2)
elif delta == 0:
  print('Tek gerçel kök vardır:')
  x = - b / 2 * a
  print (x)
else:
  print('Gerçel kök yoktur..')
```

```python
# katsayıları girilerek 2. dereceden denklemin çözümünü bulan algoritma Vs02
from math import sqrt 
a = 1
b = 2
c = 1
delta = b ** 2 - 4 * a * c

if delta > 0:
  print('İki farklı gerçel kök vardır:')
  x1 = (- b + sqrt(delta)) / 2 * a
  x2 = (- b - sqrt(delta)) / 2 * a
  print(x1 , x2)
elif delta == 0:
  print('Tek gerçel kök vardır')
  x = -b / 2 * a
  print(x)
else:
  print('Gerçel kök yoktur')
```

## <b>8. Hata Ayıklama (try...except)</b>

Geliştirdiğimiz programın çökmemesi için hataların yakalanmasını sağlarız. Bunun için try ... except yapısını kullanırız.  Ayrıca finally anahtar kelimesi kullanılarak hata olsun yada olmasın çalışması istenen kodların çalışması sağlanır.

image.png

```python
def böl(sayi1,sayi2):
  try:
    float(sayi1)
    float(sayi2)
  except:
    print('float yada int veri tipinde veri giriniz')
  else:
    try:
      float(sayi1) / float(sayi2) 
    except: 
      print('bölen sıfır olamaz')
    else:
      return float(sayi1) / float(sayi2)
  finally:
    print('Fonksiyon Çalıştı')
```

```python
böl(1,2)
```

```python
böl(1,0)
```

```python
böl(1,'a')
```

```python
böl(10,'5')
```
